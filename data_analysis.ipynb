```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
columnss = ["Number of Rows", "CSV File Size", "HDF5 File Size"]

csv_size = [79, 157, 235, 313, 392, 470, 548, 626, 705, 783]
hdf5_size = [43, 83, 122, 161, 200, 239, 278, 317, 356, 395]
num_rows = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
```

```python
df = pd.DataFrame({"Number of Rows" : num_rows, "HDF5 Size (kilobytes)" : hdf5_size, "CSV Size (kilobytes)" : csv_size})
```

```python
df
```

```python
df["HDF5 Size (kilobytes)"]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df["Number of Rows"], df["HDF5 Size (kilobytes)"], label="HDF5 Size", marker="o")
plt.plot(df["Number of Rows"], df["CSV Size (kilobytes)"], label="CSV Size", marker="o")

# Add labels, title, and legend
plt.xlabel("Number of Rows")
plt.ylabel("File Size (MB)")
plt.title("Comparison of HDF5 and CSV File Sizes")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
